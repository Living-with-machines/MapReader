## Set up download

```python
# solve issue with autocomplete
%config Completer.use_jedi = False

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
from mapreader import TileServer

my_ts = TileServer(metadata_path="../../persistent_data/metadata_OS_One_Inch_GB_WFS_light.json",
                        download_url="https://geo.nls.uk/maps/os/1inch_2nd_ed/{z}/{x}/{y}.png")
```

```python
my_ts.plot_metadata_on_map(map_extent="uk", add_text=False)
```

```python
my_ts.minmax_latlon()
```

```python
my_ts.hist_published_dates()
```


## Query maps using a list of lats/lons

```python
my_ts.query_point([55.90, -4.33])
my_ts.query_point([51.93, -5.25], append=True)
```

```python
my_ts.print_found_queries()
```

## Download map tiles

```python
my_ts.download_tileserver(mode="query", 
                               zoom_level=14,
                               pixel_closest=50,
                               output_maps_dirname="./maps")
```
