```python
from mapreader import loader
```

```python
my_files = loader("./ts_downloaded_maps/*png")
my_files.add_metadata("./ts_downloaded_maps/ts_downloaded_maps_metadata.csv")
```

```python
my_files.patchifyAll(method='meter', patch_size=500, path_save="./ts_downloaded_maps/patch_500_500/", output_format="png")
```

```python
my_files.show_sample(tree_level='patch', num_samples=12)
```
